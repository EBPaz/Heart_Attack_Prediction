```python
import pandas as pd
data_df=pd.read_csv('/content/gdrive/MyDrive/Project4/heart_attack_prediction_dataset.csv')
data_df.head()
```

```python
# Drop the un-necessary columns - "Patient ID" and "BMI"
updated_data_df = data_df.drop(columns =['Patient ID','BMI'])
updated_data_df
```

```python
# Generate our categorical variable lists
updated_df_cat = updated_data_df.dtypes[updated_data_df.dtypes == "object"].index.tolist()
```

```python
# Check the number of unique values in each column
updated_data_df[updated_df_cat].nunique()
```

```python
# Split blood pressure
updated_data_df[['systole','diastole']] = updated_data_df['Blood Pressure'].str.split("/",expand=True)
updated_data_df.head()
```

```python
# Drop original Blood Pressure column
updated_data_df = updated_data_df.drop(columns =['Blood Pressure'])
updated_data_df
```

```python
updated_data_df[['systole','diastole']] = updated_data_df[['systole','diastole']].astype(int)
```

```python
bp = []
for index, row in updated_data_df.iterrows():

    if((row["systole"] < 120) & (row["diastole"] < 80)):
     bp.append("Normal Blood Pressure")

    elif ((row["systole"] >= 120 & row["systole"] < 130) & (row["diastole"] < 80)):
     bp.append("Elevated Blood Pressure")

    elif ((row["systole"] >= 130 & row["systole"] < 140) | (row["diastole"]>=80 & row["diastole"]<=89)):
     bp.append("First Stage High Blood Pressure")

    else:
     bp.append("Second Stage High Blood Pressure")

bp_updated_data_df = updated_data_df.assign (bp = bp)
bp_updated_data_df.head()
```

```python
# Drop the un-necessary columns - "systole" and "diastole"
bp_updated_data_df = bp_updated_data_df.drop(columns =["systole","diastole"])
bp_updated_data_df
```

```python
# Generate our categorical variable lists
bp_updated_df_cat = bp_updated_data_df.dtypes[bp_updated_data_df.dtypes == "object"].index.tolist()
```

```python
# Check the number of unique values in each column
bp_updated_data_df[bp_updated_df_cat].nunique()
```

```python
bp_updated_data_df["bp"].unique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(bp_updated_data_df[bp_updated_df_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names_out(bp_updated_df_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
bp_updated_data_df = bp_updated_data_df.merge(encode_df,left_index=True, right_index=True)
bp_updated_data_df = bp_updated_data_df.drop(bp_updated_df_cat,1)
bp_updated_data_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = bp_updated_data_df["Heart Attack Risk"].values
X = bp_updated_data_df.drop(["Heart Attack Risk"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create scaler instance
X_scaler = skl.preprocessing.StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh','sigmoid'])

    # Allow kerastuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=4,
        max_value=8,
        step=2), activation=activation, input_dim=53))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 4)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=4,
            max_value=8,
            step=2),
            activation=activation))

    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])

    return nn_model
```

```python
# Import the kerastuner library
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2)
```

```python
# Run the kerastuner search for best hyperparameters
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

```python
# Get best model hyperparameters
best_hyper = tuner.get_best_hyperparameters(1)[0]
best_hyper.values
```

```python
# Evaluate best model against full test data
best_model = tuner.get_best_models(1)[0]
model_loss, model_accuracy = best_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
